# Run Firefly in an HPC facility like Sciama

Here we describe how to make use of the scripts in '../bin/SDSS' to run Firefly in an HPC facility like [Sciama](http://www.sciama.icg.port.ac.uk/).

The first step to run Firefly is to set the enviromental variables so Python finds the code and the Stellar Population Models. 

Example for a .bashrc file:
```
$FF_DIR='[your path to Firefly]/firefly_code'
$PYTHONPATH=‘${FF_DIR}/python:$PYTHONPATH’
$STELLARPOPMODELS_DIR=‘${FF_DIR}/data/stellar_population_models/'
```

The following script has two purposes:
${FF_DIR}/bin/SDSS/write_scripts_submit.py

* If 'write_scripts = True', it will write submission scripts stored in adirectory that is not tracked by git:

```
${FF_DIR}/jobs
```

2) If 'submit_scripts = True', then the 'run_stellerpop.py' (which is equivalent to the one_spectra.py) is called and it will submit the scripts and the output will be stored in a directory that is not tracked by git:
${FF_DIR}/output
